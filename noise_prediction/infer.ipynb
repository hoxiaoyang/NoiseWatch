```python
import joblib
import os
import numpy as np

from inference_utils import process_unstructured_data_to_csv, fourier_transform, extract_spectral_features
```

# Load model

```python
# Get current working directory
cwd = os.getcwd()

# Load from joblib
model = joblib.load(os.path.join(cwd, "model.joblib"))
```

# Process each frame

```python
# Sample data
# dictionary: Data should be in following format: 
# {   
#     "start_time": <timestamp>,
#     "house_id": <id>,
#     "data": [ {"timestamp": <timestamp1>, "analog_value": <value1>}, {"timestamp": <timestamp2>, "analog_value": <value2>}, ... for 30 sets of values]
# }
# time_interval: Time interval between data points in milliseconds.

sample_data = {
    "start_time": 1695763200000,
    "house_id": "house_123",
    "data": [ {"timestamp": 1695763200000 + i*10, "analog_value": np.sin(2 * np.pi * 440 * (i*10)/1000) + 0.5 * np.random.randn()} for i in range(3000)]
}
```

```python
df = process_unstructured_data_to_csv(sample_data)
```

```python
freq_df, sampling_rate = fourier_transform(df)
```

```python
features = extract_spectral_features(freq_df)
```

# Example inference

```python
prediction = model.predict([features])
```

```python
print("Predicted label index:", prediction[0])
```
